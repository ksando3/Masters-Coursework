```python
## I'm going to create a new dataframe that has just the information and columns that i will use for my analysis.
## I want to use the same variables I used in D208

default_data_RF = default_data.drop(columns = ['Population', 'CaseOrder', 'Customer_id','Interaction', 'UID', 'City', 'State',
                                                'County', 'Zip', 'Lat', 'Lng', 'TimeZone', 'Job', 'Marital', 'Tenure', 
                                               'PaymentMethod','Item1', 'Item2', 'Item3', 'Item4',
                                               'Item5', 'Item6', 'Item7', 'Item8'])
```

```python
## Checking the new data set. 
## or to check on a specific customer without any data loss. Should have 18 columns now.
default_data_RF.head()
```

```python
## Getting a list of all of the columns names except for the CaseOrder and Customer_id

churned_cont =['Children', 'Age', 'Income', 'Outage_sec_perweek', 'Contacts', 'Email', 'Yearly_equip_failure', 'MonthlyCharge', 'Bandwidth_GB_Year']
churned_cat = ['Churn', 'Techie', 'Port_modem', 'Tablet', 'Phone', 'Multiple', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling']
churned_misc = ['Area', 'Gender', 'InternetService', 'Contract']
```

```python
print(default_data_RF.columns.tolist())
```

```python
## Dr. Elleh video lecture 'D209 T1 Building KNN in Python Med'
for i in range(0, len(churned_misc)):
    a, b, c = default_data_RF[churned_misc[i]].unique()
    default_data_RF[churned_misc[i]] = default_data_RF[churned_misc[i]].str.replace(' ', '')
    encoded = pd.get_dummies(default_data_RF[churned_misc[i]], prefix = churned_misc[i])
    frames = [default_data_RF, encoded]
    default_data_RF = pd.concat(frames, axis = 1)
    default_data_RF = default_data_RF.drop([churned_misc[i]], axis = 1)
default_data_RF.head()
```

```python
## Replacing all Yes/No responses with an int value corresponding to Yes = 1 and No = 0.
## Per my code from D206

default_data_RF = default_data_RF.replace({"Yes": 1, "No": 0})
#default_data_RF.head()
```

```python
seaborn.heatmap(data = default_data_RF[churned_cat].corr(), annot = True, cmap = 'coolwarm')
```

```python
default_data_RF.to_csv('C:/Users/cfman/OneDrive/Desktop/WGUClasses/D603MachineLearning/Task1/churn_data_RF.csv')
#print(default_data_RF.head(5))
```

```python
train, validate, test = np.split(default_data_RF.sample(frac=1, random_state=42),
                                 [int(.6*len(default_data_RF)), int(.8*len(default_data_RF))])
```

```python
ex_train = train.drop(['Churn'], axis = 1)
ex_val = validate.drop(['Churn'], axis = 1)
ex_test = test.drop(['Churn'], axis = 1)

y_train = train['Churn']
y_val = validate['Churn']
y_test = test['Churn']
```

```python
print("Is ex_train 60% of the full set: ", len(ex_train) == round(len(default_data_RF) * 0.6))
print("Is ex_val 20% of the full set: ", len(ex_val) == round(len(default_data_RF) * 0.2))
print("Is ex_test 20% of the full set: ", len(ex_test) == round(len(default_data_RF) * 0.2))
print("Is y_train 60% of the full set: ", len(y_train) == round(len(default_data_RF) * 0.6))
print("Is y_test 20% of the full set: ", len(y_test) == round(len(default_data_RF) * 0.2))
print("Is y_val 20% of the full set: ", len(y_val) == round(len(default_data_RF) * 0.2))
```

```python
ex_train.to_csv('C:/Users/cfman/OneDrive/Desktop/WGUClasses/D603MachineLearning/Task1/ex_train.csv')
ex_val.to_csv('C:/Users/cfman/OneDrive/Desktop/WGUClasses/D603MachineLearning/Task1/ex_val.csv')
ex_test.to_csv('C:/Users/cfman/OneDrive/Desktop/WGUClasses/D603MachineLearning/Task1/ex_test.csv')
y_train.to_csv('C:/Users/cfman/OneDrive/Desktop/WGUClasses/D603MachineLearning/Task1/y_train.csv')
y_val.to_csv('C:/Users/cfman/OneDrive/Desktop/WGUClasses/D603MachineLearning/Task1/y_val.csv')
y_test.to_csv('C:/Users/cfman/OneDrive/Desktop/WGUClasses/D603MachineLearning/Task1/y_test.csv')
```

```python
rfr = RandomForestClassifier()
rfr = rfr.fit(ex_train, y_train)
y_pred = rfr.predict(ex_test.values)
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(cnf_matrix)
disp.plot()
plt.show()
print(classification_report(y_test, y_pred))
```

```python
## Initial ROC Curve
probs = rfr.predict_proba(ex_test.values)[:,1]
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, probs)

plt.plot([0,1], [0,1], 'k--')
plt.plot(false_positive_rate, true_positive_rate, label = 'KNN')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Random Forest ROC Curve')
plt.show()
```

```python
## Initial AUC
auc = roc_auc_score(y_test, probs)
print("AUC: ", auc)
```

```python
## Dr. Elleh's video lecture 'D209 Webinar Splitting Data and Creating Model'
# Determining the best parameters while using the validation dataset
rfr_param = RandomForestClassifier(random_state = 1)
param_grid = {'n_estimators': [10,50,100], 'max_features': [2,3,4], 
              'max_depth' : [8, None]}

# Defining parameter range
grid = GridSearchCV(rfr_param, param_grid, cv = 10)

# Fitting the model for grid search
grid_search = grid.fit(ex_val, y_val)

print('The best parameters are: ', grid_search.best_params_)
```

```python
## Using the test data set with the given parameters. Predicting on the validation set.
rfr_update = RandomForestClassifier(max_depth = None, max_features = 4, n_estimators = 50, random_state = 1)
rfr_update = rfr_update.fit(ex_test, y_test)
y_pred_update = rfr_update.predict(ex_val.values)
```

```python
cnf_matrix_update = metrics.confusion_matrix(y_val, y_pred_update)
disp_update = ConfusionMatrixDisplay(cnf_matrix_update)
disp_update.plot()
plt.show()
print(classification_report(y_val, y_pred_update))
```

```python
## Updated ROC Curve
probs_update = rfr.predict_proba(ex_val.values)[:,1]
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_val, probs_update)

plt.plot([0,1], [0,1], 'k--')
plt.plot(false_positive_rate, true_positive_rate, label = 'KNN')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Random Forest ROC Curve')
plt.show()
```

```python
## Initial AUC
auc = roc_auc_score(y_val, probs_update)
print("AUC: ", auc)
```
