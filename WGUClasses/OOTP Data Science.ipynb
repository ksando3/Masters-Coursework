```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
os.chdir(r'C:\Users\cfman\OneDrive\Desktop\OOTP Data Science\Batting\All Live')
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
combined_csv_batters = pd.concat([pd.read_csv(f) for f in all_filenames])
```

```python
combined_csv_batters.fillna(0, inplace = True)
```

```python
wOBAweighted = (combined_csv_batters['wOBA'] * combined_csv_batters['AB']) / sum(combined_csv_batters['AB']).groupby('Name')

k = {'G':'sum', 'PA':'sum', 'AB':'sum', 'H':'sum', '2B':'sum', '3B':'sum', 'HR':'sum', 'RBI':'sum', 'R':'sum', 'BB':'sum',
    'HP':'sum', 'SO':'sum', 'RC':'sum', 'WAR':'sum', 'SB':'sum', 'CS':'sum', 'BatR':'sum'} 
edited_csv = combined_csv_batters.groupby('Name').agg(k)
```

```python
edited_csv['1B'] = edited_csv['H'] - edited_csv['2B'] - edited_csv['3B'] - edited_csv['HR']
edited_csv['AVG'] = round(edited_csv['H'] / edited_csv['AB'] , 3)
edited_csv['OBP'] = round((edited_csv['H'] + edited_csv['BB'] + edited_csv['HP']) / edited_csv['PA'] , 3)
edited_csv['SLG'] = round((edited_csv['1B'] + 2*edited_csv['2B'] + 3*edited_csv['3B'] + 4*edited_csv['HR']) / edited_csv['AB'], 3)
edited_csv['ISO'] = round(edited_csv['SLG'] - edited_csv['AVG'], 3)
edited_csv['OPS'] = round(edited_csv['OBP'] + edited_csv['SLG'], 3)
```

```python
edited_csv.sort_values('WAR', ascending = False)
```
