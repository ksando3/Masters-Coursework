```python
## Dr. Elleh video lecture 'D209 T1 Building KNN in Python Med'
for i in range(0, len(churned_misc)):
    a, b, c = churned_data_RF[churned_misc[i]].unique()
    churned_data_RF[churned_misc[i]] = churned_data_RF[churned_misc[i]].str.replace(' ', '')
    encoded = pd.get_dummies(churned_data_RF[churned_misc[i]], prefix = churned_misc[i])
    frames = [churned_data_RF, encoded]
    churned_data_RF = pd.concat(frames, axis = 1)
    churned_data_RF = churned_data_RF.drop([churned_misc[i]], axis = 1)
churned_data_RF.head()
```

```python
## From https://campus.datacamp.com/courses/intermediate-regression-with-statsmodels-in-python/multiple-logistic-regression-4?ex=3
## https://www.statsmodels.org/dev/generated/statsmodels.discrete.discrete_model.Logit.predict.html#statsmodels.discrete.discrete_model.Logit.predict-parameters

explanatory = churned_data_RF.drop(['Tenure'], axis = 1)
target = churned_data_RF['Tenure']

explanatory_train,explanatory_test,target_train,target_test = train_test_split(explanatory,target,test_size=0.2,random_state=1) 

explanatory_train.to_csv('C:/Users/cfman/OneDrive/Desktop/WGUClasses/D209 Data Mining 1/Task 2/explanatory_train.csv')
explanatory_test.to_csv('C:/Users/cfman/OneDrive/Desktop/WGUClasses/D209 Data Mining 1/Task 2/explanatory_test.csv')
target_train.to_csv('C:/Users/cfman/OneDrive/Desktop/WGUClasses/D209 Data Mining 1/Task 2/target_train.csv')
target_test.to_csv('C:/Users/cfman/OneDrive/Desktop/WGUClasses/D209 Data Mining 1/Task 2/target_test.csv')
```

```python
## Dr. Elleh's video lecture 'D209 Webinar Splitting Data and Creating Model'
# Determining the best parameters
rfr = RandomForestRegressor(random_state = 1)
param_grid = {'n_estimators': [10,50,100], 'max_features': [2,3,4], 
              'max_depth' : [8, None]}

# Defining parameter range
grid = GridSearchCV(rfr, param_grid, cv = 10)

# Fitting the model for grid search
grid_search = grid.fit(explanatory_train, target_train)

print('The best parameters are: ', grid_search.best_params_)

print('Training Score (MSE): ', grid_search.best_score_)
print('Training Score (RMSE): ', (grid_search.best_score_)**(1/2))
y_train_pred = grid_search.predict(explanatory_train)
print("Training - R squared score for the model: ", r2_score(target_train, y_train_pred))
```

```python
rfr = RandomForestRegressor(n_estimators = 100, max_features = 4, max_depth = None, random_state = 1)
rfr = rfr.fit(explanatory_train, target_train)
y_pred = rfr.predict(explanatory_test)

print("Testing - Mean squared error for the model: ", mean_squared_error(target_test, y_pred))
print("Testing - Root Mean squared error for the model: ", mean_squared_error(target_test, y_pred)**(1/2))
print("Testing - R squared for the model: ", r2_score(target_test, y_pred))
```

```python
importances = pd.Series(data = grid_search.best_estimator_.feature_importances_, index = explanatory_train.columns)

importances_sorted = importances.sort_values()

importances_sorted.plot(kind = 'barh', color = 'lightgreen')
plt.title('Feature Importances')
plt.show()
```
